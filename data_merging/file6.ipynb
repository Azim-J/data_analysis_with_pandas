```python
import pandas as pd
lotr1 = pd.read_csv("LOTR.csv")
lotr2 = pd.read_csv("LOTR_2.csv")
# A simple combination of the two tables
lotr_combined=pd.concat([lotr1, lotr2])
lotr_combined
```

```python
#Only the entrys that are in both tables
lotr_inner=pd.merge(left=lotr1, right=lotr2, how='inner', on=['FellowshipID', "FirstName"])
lotr_inner
```

```python
#All the entries in either one of the two tables
lotr_outer=pd.merge(left=lotr1, right=lotr2, how='outer')
lotr_outer
```

```python
#Only the rows in the right table, with data added from the left table (The skills column)
lotr_right=pd.merge(left=lotr1, right=lotr2, how='right')
lotr_right
```

```python
#Creating pairs of people
lotr_cross=pd.merge(left=lotr1, right=lotr2, how='cross')
lotr_cross
```

```python
#The two tables combined into one, with the same titled columns differentiated
lotr_join=lotr1.join(lotr2, lsuffix='_1', rsuffix="_2")
lotr_join
```
