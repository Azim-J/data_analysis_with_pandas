```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('world_population.csv')
#Sorting and Indexing by range of ranks
upper_limit=20 #The upper limit of the range of ranks (the smaller number)
lower_limit=25 #The lower limit of the range of ranks (the bigger number)
dfranks = df[(df["Rank"] <= lower_limit) & (df["Rank"] >= upper_limit)].sort_values(by="Rank", ascending=True).set_index("Rank")
dfranks
```

```python
#Sorting and indexing by continent
continents="Asia Africa" #The list of continents in which the countries  are located, seperated by spaces
continents=continents.split()
dfcontinents = df[df["Continent"].isin(continents)].sort_values(by="Continent").set_index("Continent")
dfcontinents
```

```python
countries_list="Canada Bulgaria Fiji Finland Guam Laos Malta Mali" #Enter the list of countries you would like to see, seperated by spaces
countries_list=countries_list.split()
dfcountries = df[df["Country"].isin(countries_list)].sort_values(by="Country").set_index("Country")
dfcountries
```
